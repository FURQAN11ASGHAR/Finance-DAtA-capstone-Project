# Gatthering stocks data of the banks from 2006-2016

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = yf.download('BAC', start=start, end=end, auto_adjust=True)       #bankof America
C = yf.download('C', start=start, end=end, auto_adjust=True)           # Citigroup
GS = yf.download('GS', start=start, end=end, auto_adjust=True)         # Gold man Sachs
JPM = yf.download('JPM', start=start, end=end, auto_adjust=True)       # JPMorgan Chase
MS = yf.download('MS', start=start, end=end, auto_adjust=True)         # Morgan Stanley
WFC = yf.download('WFC', start=start, end=end, auto_adjust=True)       # Wells fargo Bank
```

```python
GS
```

# Restructuring data according to ease Data analysis

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
Bank_stocks=bank_stocks.droplevel('Ticker',axis=1)
Bank_stocks.head()
```

```python
Bank_stocks.columns.set_names('Stock info', level=1, inplace=True)
Bank_stocks.columns.set_names('Bank',level=0, inplace=True)
```

```python
Bank_stocks.head()
```
